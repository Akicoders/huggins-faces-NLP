```python
from transformers import pipeline

# Analisis de Sentimientos
classifier_en = pipeline("sentiment-analysis")
# Bueno primera prueba es que sabemos es que todo va a en ingles(bueno en el salida nos dice).
print(classifier_en(["A mi me gusta mucho la inteligencia artificial y la computacion cuantica.", "I love AI and Quamtum", "I hate my T-shirt"]))
# Probamo con uno es español
classifier_es = pipeline("sentiment-analysis", model="tabularisai/multilingual-sentiment-analysis")
print(classifier_es(["A mi me gusta mucho la inteligencia artificial y la computacion cuantica.", "I love AI and Quamtum", "I hate my T-shirt"]))
```

```python
# Zero Shot Classification
zsc = pipeline("zero-shot-classification")
# Creo que lo puse muy dificil para este modelo
zsc("In my day I used to spend three hours learning and searching for information, always programming in Java, Javascript and Python. I love this, but this day I decided to learn NLP with Hugging Faces, it's fantastic.",
    candidate_labels=["computing", "business", "education"])
# Probemos otro (va a hacer un poco mas complejo pero es por nuestro bien)
# Los modelos siempre no van a estar con solo poner el nombre y ya esta, hay unos que tienen otros procedimientos y hasta sus propia documentacion.
# Asi que vamos a usar este:
# https://huggingface.co/knowledgator/gliclass-modern-base-v2.0-init
```

```python
from gliclass import GLiClassModel, ZeroShotClassificationPipeline
from transformers import AutoTokenizer
# en la documentacion hay mas modelos pero utilizamos el mejor
model = GLiClassModel.from_pretrained("knowledgator/gliclass-modern-base-v2.0-init")
tokenizer = AutoTokenizer.from_pretrained("knowledgator/gliclass-modern-base-v2.0-init")
pipeline = ZeroShotClassificationPipeline(model, tokenizer, classification_type='multi-label', device='cuda:0')

text = "In my day I used to spend three hours learning and searching for information, always programming in Java, Javascript and Python. I love this, but this day I decided to learn NLP with Hugging Faces, it's fantastic."
labels = ["computing", "business","education"]
#pipeline :D
# threshold = Segun el porcentaje como minimo tener sino no lo coloca
results = pipeline(text, labels, threshold=0.5)[0] # hay un solo texto por eso [0]
print(results)
```

```python

# Text-Generation
# No es muy bueno el por defecto, pero podemos añadirle cosas
tg = pipeline("text-generation")
print(tg("I love programing , my favorite programing language is Python but ", max_length=30,num_return_sequences=2,))

# Podriamos usar Deep Seek R1 (PESA COMO 60 GB 💀), asi que use el mas pequeño aver que tal,
# Recomiendo probarlo aparte si usas google colab porque se me cuelga porq la ram no aguanta :c
tgDS = pipeline("text-generation", model="deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B")
print(tgDS("I love programing , my favorite programing language is Python but "))
```

```python
summarizer= pipeline("summarization")
resumen =  summarizer("""
  Quantum Computing: Essential Concepts and Principles

Quantum computing is a revolutionary field of study that leverages the principles of quantum mechanics to perform computations that are infeasible for classical computers. Unlike classical bits, which can be either 0 or 1, quantum bits (qubits) can exist in multiple states simultaneously due to the phenomena of superposition and entanglement.

1. Qubits: The Fundamental Unit

A qubit is the basic unit of quantum information. Unlike classical bits, qubits leverage quantum mechanical properties to exist in a superposition of states. This means that a qubit can be in a state of both 0 and 1 simultaneously, increasing computational power exponentially.

2. Superposition

Superposition is a principle that allows quantum systems to be in multiple states at once. In quantum computing, this means that a qubit can represent both 0 and 1 at the same time until it is measured. This characteristic enables quantum computers to process a vast number of possibilities in parallel.

3. Entanglement

Quantum entanglement is a phenomenon where two or more qubits become interconnected, meaning the state of one qubit directly affects the state of another, regardless of the distance between them. Entanglement enables faster and more secure communication and plays a crucial role in quantum algorithms.

4. Quantum Gates and Circuits

Quantum gates are operations that manipulate qubits, similar to logic gates in classical computing. Examples of quantum gates include:

Hadamard Gate (H): Creates superposition by transforming a qubit into an equal probability state of 0 and 1.

Pauli-X, Y, Z Gates: Perform rotations around different axes in the Bloch sphere, changing the qubit state.

CNOT (Controlled NOT) Gate: A two-qubit gate used for entanglement and error correction.

Toffoli Gate: A universal gate for reversible computation.
Quantum circuits are composed of multiple quantum gates applied in sequence to perform computations.

5. Quantum Measurement

Quantum measurement collapses a qubit's superposition state into a definite classical value (either 0 or 1). The act of measurement is probabilistic, meaning that the output depends on the qubit's quantum state before measurement.

6. Quantum Algorithms

Quantum algorithms leverage quantum mechanical properties to solve problems more efficiently than classical algorithms. Some notable quantum algorithms include:

Shor’s Algorithm: Efficiently factors large numbers, threatening current cryptographic systems.

Grover’s Algorithm: Provides a quadratic speedup for searching an unsorted database.

Quantum Fourier Transform (QFT): A key component in many quantum algorithms, particularly for signal processing and cryptographic applications.

7. Quantum Error Correction

Due to the fragile nature of quantum states, quantum error correction techniques are necessary to preserve information and ensure accurate computations. Methods such as the Shor Code and Surface Code help mitigate decoherence and quantum noise.

8. Quantum Hardware and Technologies

Various technologies are being explored to build scalable quantum computers, including:

Superconducting Qubits: Used by companies like IBM and Google, these qubits are based on Josephson junctions.

Trapped Ions: Utilize ions trapped in electromagnetic fields and manipulated using lasers.

Topological Qubits: A promising approach that aims to reduce errors using braiding properties of quantum states.

9. Quantum Supremacy and Practical Applications

Quantum supremacy refers to the point where a quantum computer can solve a problem faster than the best classical supercomputers. Google achieved quantum supremacy in 2019 by demonstrating a computation that would take classical supercomputers thousands of years. Potential applications of quantum computing include:

Cryptography: Breaking traditional encryption and developing quantum-safe cryptographic methods.

Drug Discovery: Simulating molecular interactions for pharmaceutical advancements.

Optimization Problems: Enhancing logistics, finance, and supply chain management.

Artificial Intelligence: Speeding up machine learning processes through quantum-enhanced algorithms.

10. Challenges and Future Prospects

Despite its potential, quantum computing faces several challenges:

Scalability: Building large-scale, error-free quantum systems.

Decoherence and Noise: Maintaining qubit stability over time.

High Operational Costs: Quantum systems require extreme cooling and specialized environments.
As research progresses, hybrid quantum-classical systems are expected to emerge, making quantum computing more accessible and practical.
""")
resumen_texto= resumen[0]['summary_text']
print(resumen_texto)
```
